```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_df=pd.read_csv(city_data_to_load)
ride_df=pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
combined_df=pd.merge(ride_df, city_df, how="left", on=["city","city"])

# Display the data table for preview

combined_df.head()

```

```python
# Obtain the x and y coordinates for each of the three city types
urban_df=combined_df.loc[combined_df["type"]=="Urban"]
suburban_df=combined_df.loc[combined_df["type"]=="Suburban"]
rural_df=combined_df.loc[combined_df["type"]=="Rural"]


# #total ride per city (urban)
x_axis_urban = urban_df.groupby(["city"]).count()["ride_id"]
#Average fare per city (urban)
y_axis_urban = urban_df.groupby(["city"]).mean()["fare"]
#Driver Count (urban)
s_axis_urban = urban_df.groupby(["city"]).mean()["driver_count"]


#total ride per city (suburban)
x_axis_suburban = suburban_df.groupby(["city"]).count()["ride_id"]
#Average fare per city (urban)
y_axis_suburban = suburban_df.groupby(["city"]).mean()["fare"]
#Driver Count (urban)
s_axis_suburban = suburban_df.groupby(["city"]).mean()["driver_count"]


#total ride per city (rural)
x_axis_rural = rural_df.groupby(["city"]).count()["ride_id"]
#Average fare per city (rural)
y_axis_rural = rural_df.groupby(["city"]).mean()["fare"]
#Driver Count (rural)
s_axis_rural = rural_df.groupby(["city"]).mean()["driver_count"]

# Incorporate the other graph properties
#Chart Title
plt.title("Pyber Ride Sharing Data 2016")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare($)")


#size of plots


#scatter plots
urban=plt.scatter(x_axis_urban, y_axis_urban, s_axis_urban, alpha=.5, color="r", edgecolor="black", label="Urban")
suburban=plt.scatter(x_axis_suburban, y_axis_suburban, s_axis_suburban, alpha=.5, edgecolor="black", color="b", label="Suburban")
rural=plt.scatter(x_axis_rural, y_axis_rural,s_axis_rural, alpha=.5, color="y", edgecolor="black", label="Rural")

# Create a legend
plt.legend(handles=[urban,suburban,rural])

#Grid
plt.grid()

# Size Legend
plt.text(45,30,"Note:Circle size correlates with driver count per city")

#print Chart

plt.show()

#save
plt.savefig("Bubble.png")
```

## Bubble Plot of Ride Sharing Data

## Total Fares by City Type

```python
# Calculate Type Percents
#% of Total Fares by City Type
urban_total_fare = urban_df.sum()["fare"]
suburban_total_fare = suburban_df.sum()["fare"]
rural_total_fare = rural_df.sum()["fare"]

#Create legend
plt.title("% of Total Fares by City Type")

#size
city_pct = [urban_total_fare, suburban_total_fare, rural_total_fare]

#Labels
city_label = ["Urban", "Suburban", "Rural"]

#Color
colors = ["lightcoral", "skyblue", "yellow"]

#Explode
explode = (0.1,0,0)

# Build Pie Chart
plt.pie(city_pct, labels=city_label, explode=explode, colors=colors, counterclock=False, autopct="%1.1f%%", shadow=True, startangle=140)

# Show Figure
plt.show()

# Save Figure
#save
plt.savefig("FarebyCity.png")
```

## Total Rides by City Type

```python
# Calculate Ride Percents
urban_total_rides = urban_df.count()["ride_id"]
suburban_total_rides = suburban_df.count()["ride_id"]
rural_total_rides = rural_df.count()["ride_id"]

#Create legend
plt.title("% of Total Rides by City Type")

#size
city_pct = [urban_total_rides, suburban_total_rides, rural_total_rides]

#Labels
city_label = ["Urban", "Suburban", "Rural"]

#Color
colors = ["lightcoral", "skyblue", "yellow"]

#Explode
explode = (0.1,0,0)

# Build Pie Chart
plt.pie(city_pct, labels=city_label, explode=explode, colors=colors, counterclock=False, autopct="%1.1f%%", shadow=True, startangle=140)

# Show Figure
plt.show()

# Save Figure
plt.savefig("RidesbyCity.png")
```
