```python
#dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
```

```python
#create variable to hold csv file
file_path = "baseball_project.csv"
```

```python
# Read our Data file with the pandas library
base_df=pd.read_csv(file_path)
# base_df = base_df.set_index("yearID")
base_df.head()
```

```python
base_df = base_df[base_df["yearID"]>=2010]
hit = base_df.groupby("yearID").mean()["H"]
hit = pd.DataFrame(hit)
hit
```

```python
#Hits Average per year
```

```python
x_axis = hit.index
y_axis = hit["H"]
plt.bar(x_axis,y_axis)
```

```python
#ERA Average per year
date_range=base_df[base_df["yearID"]==2010]
ERA_mean_2010=date_range["ERA"].mean()

date_range=base_df[base_df["yearID"]==2011]
ERA_mean_2011=date_range["ERA"].mean()

date_range=base_df[base_df["yearID"]==2012]
ERA_mean_2012=date_range["ERA"].mean()

date_range=base_df[base_df["yearID"]==2013]
ERA_mean_2013=date_range["ERA"].mean()

date_range=base_df[base_df["yearID"]==2014]
ERA_mean_2014=date_range["ERA"].mean()

date_range=base_df[base_df["yearID"]==2015]
ERA_mean_2015=date_range["ERA"].mean()

date_range=base_df[base_df["yearID"]==2016]
ERA_mean_2016=date_range["ERA"].mean()

date_range=base_df[base_df["yearID"]==2017]
ERA_mean_2017=date_range["ERA"].mean()

print(ERA_mean_2010,ERA_mean_2011,ERA_mean_2012,ERA_mean_2013,ERA_mean_2014,ERA_mean_2015,ERA_mean_2016,ERA_mean_2017)
```
